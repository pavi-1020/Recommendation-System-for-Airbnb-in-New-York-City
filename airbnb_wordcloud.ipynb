```python
import pandas as pd
from IPython.display import Image, HTML
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
ab = pd.read_csv("airbnb_clean2.csv",encoding = 'unicode_escape',usecols = ['id', 'name'])
ab.head(10)
```

```python
ab['name'] = ab['name'].astype('str')
name_corpus = ' '.join(ab['name'])
```

```python
name_wordcloud = WordCloud(stopwords = STOPWORDS, background_color = 'white', height = 2000, width = 4000).generate(name_corpus)
plt.figure(figsize = (16,8))
plt.imshow(name_wordcloud)
plt.axis('off')
plt.show()
```
